```python
from typing import List
from collections import deque, defaultdict, Counter
```

```python
def add_item(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items
```

```python
add_item("apple")
# ['apple']

add_item("banana")
# ['apple', 'banana']  ❌ unexpected
```
